```python
dfo = pd.read_csv('train.csv', low_memory=False)
dfo.dropna()
dfo.head()
```

```python
df = dfo.set_index(['Store', 'Date']).sort_index()
df.head()
```

### Stacking and Unstacking

You've seen this briefly with GroupBy results. Let's go into more detail now. Stacking however only works if there aren't duplicates in your index. It must be uniquely specifying the row.

```python
df.unstack()
```

```python
df.Sales.unstack()
```

```python
df.Sales.unstack(level='Store')
```

What happens if you have more levels?

```python
df2 = dfo.set_index(['Store', 'Date', 'Open']).sort_index()
df2.head()
```

```python
df2.Customers.unstack(level='Open')
```

```python
tmp = df2.Customers.unstack(level=['Store','Open'])
tmp
```

```python
tmp.index
```

```python
tmp.columns
```

```python
tmp.stack(future_stack=True)
```

```python
tmp.stack(future_stack=True, level=['Store', 'Open']).to_frame()
```

```python
tmp.stack(future_stack=True).stack()
```

### Motivational Example
Dammit Rizal, I need to see the yearly trend over each store!

```python
dfo.head()
```

```python
x = dfo.set_index(['Date'])[['Sales']]
x
```

```python
x2 = x.unstack()
```

```python
means = x2.mean(numeric_only=True, axis=1)
x_norm = x2.div(means, axis=0)
```

```python
x_norm
```

```python
x2.mean(numeric_only=True, axis=1)
```

```python
fig, ax = plt.subplots(figsize=(13, 6))
sb.heatmap(data=x_norm, ax=ax);
```
