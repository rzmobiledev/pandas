# Saving and Serialising a Dataframe

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.random(size=(100000, 4)), columns=['A','B','C','D'])
df.head()
```

```python
df.to_csv('save.csv', index=False, float_format='%0.4f')
```

```python
df.to_pickle('save.pkl')
```
