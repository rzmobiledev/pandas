# Merging Types

Everything we did in the previous example is an example of an inner join. That is, you get a row in the input if you have a row on both the left and right dataframes.

```python
import pandas as pd

df_s = pd.read_csv(r'students.csv')
df_g = pd.read_csv(r'grades1.csv')
df_p = pd.read_csv(r'contact.csv')
```

```python
df_g.tail()
```

```python
df_desn = df_g[df_g.course == 'DESN101']
print(df_desn.shape, df_s.shape)
```

So we have 25 students, but only 12 of them took design 101.

## Merge types

```python
pd.merge(df_s, df_desn, left_on='id', right_on='student_id')
```

```python
pd.merge(df_s, df_desn, left_on='id', right_on='student_id', how='inner') 
```

```python
pd.merge(df_s, df_desn, left_on='id', right_on='student_id', how='left')
```

Let's make a small custom dataset to ilustrate an outer join.

```python
df_a = pd.DataFrame({'A': ['x','y','z'], 'B': [1, 2, 3]})
df_b = pd.DataFrame({'A': ['u','v', 'x'], 'C': [5.0, 4.0, 3.0]})

display(df_a, df_b)
```

```python
pd.merge(df_a, df_b, on='A')
```

```python
pd.merge(df_a, df_b, on='A', how='outer')
```

## Duplicate keys

```python
df_c = pd.DataFrame({'A': ['x','x','z'], 'B': [1, 2, 3]})
df_d = pd.DataFrame({'A': ['u','x', 'x'], 'C': [5.0, 4.0, 3.0]})

display(df_c, df_d)
```

```python
pd.merge(df_c, df_d, on='A')
```

As you can see, merging two columns with duplicated join keys can easily baloon out how many rows you have.

## Duplicate columns

```python
df_e = pd.DataFrame({'A': ['x','y','z'], 'B': [1, 2, 3]})
df_f = pd.DataFrame({'A': ['u','v', 'x'], 'C': [5.0, 4.0, 3.0]})

display(df_e, df_f)
```

```python
pd.merge(df_e, df_f, on='A')
```

## Composite Keys
Sometimes one column is just not enough

```python
df_1 = pd.DataFrame({
    'year': [2000, 2000, 2001, 2001],
    'sem': [1, 2, 1, 2],
    'fee': [200, 200, 200, 200]
})

df_2 = pd.DataFrame({
    'year': [2000, 2000, 2001, 2001],
    'sem': [1, 2, 1, 2],
    'student': [1, 2, 2, 3],
    'discount': [0.1, 0.2, 0.2, 1.0]
})

df_3 = pd.DataFrame({
    'student': [1, 2, 3, 4, 5]
})

display(df_1, df_2, df_3)

from IPython.display import Javascript
Javascript('this.element.attr("style", "flex-direction: row;")')
```

So we have some nominal fees per semester, but some students are on a scholarship. We want to calculate the amount due for each student. Let's start with just the students that have a discount.

```python
combined = pd.merge(df_1, df_2, on=['year', 'sem'])
combined['due'] = combined.fee * (1 - combined.discount)
combined
```

```python
pd.merge(df_3, df_2, on='student', how='left')
```

Notice the problem here? Even though discount could be filled to 0, we don't have a year and semester join on. So what we want is a new dataframe which has MxN rows for M students and N year+sem groups. This is known as a Cartesian product, or a cross join. There are there ways to do this :
- pd.MultiIndex.from_product plus reshape
- pd.core.reshape.util.cartesian_product plus constructing the DataFrame.
- Adding a useless column and then removing it.

```python
df_1['key'], df_3['key'] = 1, 1
df_cross = pd.merge(df_1, df_3, on='key')
df_cross
```

```python
df_cross.drop('key', axis=1)
```

```python
all_fees = pd.merge(df_cross, df_2, on=['student', 'year', 'sem'], how='left')
all_fees
```

```python
all_fees.discount.fillna(0)
all_fees['due'] = all_fees.fee * (1 - all_fees.discount)
all_fees
```
