```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
students = pd.read_csv(r'students.csv')
teachers = pd.read_csv(r'teachers.csv')
grades1 = pd.read_csv(r'grades1.csv')
grades2 = pd.read_csv(r'grades2.csv')
contacts = pd.read_csv(r'contact.csv')
```

```python
display(grades1.head(), grades2.head())
```

```python
grades1['semester'] = 1
grades2['semester'] = 2

grades = pd.concat([grades1, grades2]).reset_index(drop=True)
grades
```

## Mergin horizontal

```python
display(students.head(), grades.head())
```

```python
student_grades = pd.merge(students, grades, left_on='id', right_on='student_id')
student_grades
```

```python
print(students.shape, grades.shape, student_grades.shape)
```

```python
students2 = students.rename({'id': 'student_id'}, axis=1)
display(students2.head(2), contacts.head(2))
```

```python
students_full = students2.merge(contacts, on='student_id')
students_full.head()
```

```python
df = pd.merge(left=students2, right=grades, on='student_id')
df.head()
```

```python
df = df.merge(teachers, on='course')
df
```

```python
df = df.merge(contacts, on='student_id')
df
```

How many classes that student failed

```python
df2 = df.loc[df.grade == 'F', ['student_id', 'course']]
method1 = df2.student_id.value_counts().reset_index(name='count')
method1
```

```python
sids = method1[method1['count'] >= 3].student_id
```

```python
students_full[students_full.student_id.isin(sids)]
```
