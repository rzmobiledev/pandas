# Basic Plot
We'll be using Kaggle Hearth Disease UCI dataset as an example. You can find it here:
https://www.kaggle.com/ronitf/heart-disease-uci

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('heart.csv')
```

```python
df.head()
```

## Bar Plots
Comparing discrete data. Available through either pandas or matplotlib.

```python
chest_pain = df.groupby(by='cp').median().reset_index()
chest_pain
```

```python
chest_pain.plot.bar(x='cp');
```

```python
chest_pain.plot.bar(x='cp');
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label='age')
ax.set_xlabel('cp')
ax.legend();
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label='age',
edgecolor='k', color=['red','blue','green','black'])
ax.set_xlabel('cp')
ax.set_xticks(chest_pain['cp']);
```

## Seaborn

```python
ax = sb.barplot(x='cp', y='age', err_kws={'color': 'w'}, capsize=0.1, data=df)
ax;
```

## Scatter Plots

```python
df.plot.scatter('age', 'trestbps');
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df.age, y=df.trestbps, marker='o', 
s=60, c=df.age, edgecolors='k', alpha=0.5)
ax.set_xlabel('age')
ax.set_ylabel('trestbps');
```

```python
sb.scatterplot(x='age', y='trestbps', hue='age', s=30, edgecolor=None, data=df);
```

## Line Plots

```python
ages = df.groupby(by='age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age', ['chol', 'trestbps']);
```

```python
fig, ax = plt.subplots()
ax.plot(ages['age'], ages['trestbps'], ls=':', lw=2)
ax.set_xlabel('age')
ax.set_ylabel('trestbps');
```

```python
sb.lineplot(x='age', y='trestbps', data=ages);
```
