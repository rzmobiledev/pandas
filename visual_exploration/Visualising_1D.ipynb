# Visualising 1D Distribution
We'll be using the Kaggle Heart Disease UCI dataset as an example. You can find it here:
https://www.kaggle.com/ronitf/heart-disease-uci

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv(r'heart.csv')
df.head(2)
```

## Histogram
The most common of all ways to show a distribution

```python
df.age.plot.hist(bins=30);
```

```python
plt.hist(df.age, bins=30);
```

```python
fig, ax = plt.subplots()
ax.hist(df.trestbps, bins=20, histtype='step', label='trestbps')
ax.hist(df.thalach, bins=20, histtype='stepfilled', label='thalach', alpha=0.3, edgecolor='w')
ax.legend();
```

## Box Plots
Show mean, quartiles, IQR an outliers. But i don't like it

```python
df[['trestbps', 'thalach']].plot.box();
```

```python
df[['trestbps', 'thalach']].describe()
```

```python
sb.boxplot(data=df[['trestbps', 'thalach']]);
```

```python
sb.boxplot(x='cp', y='trestbps', data=df);
```

## if we use it with pandas

```python
df.groupby('cp').boxplot(column='trestbps', sharex=True, layout=(1,4), grid=False);
```

 ## Violin Plots
 Like box plots..but better. Not a staple of pandas at the moment (thought it has been requested)

```python
fig, ax = plt.subplots()
ax.violinplot(df[['trestbps', 'thalach']].to_numpy());
```

```python
fig, ax = plt.subplots()
ax.violinplot(df[['trestbps', 'thalach']].to_numpy(), bw_method=0.2);
```

```python
sb.violinplot(data=df[['trestbps', 'thalach']], inner='quartile', bw_method=0.2);
```
